```python
import numpy as np
import matplotlib.pyplot as plt

# Definir parâmetros do transistor
beta = 100
Vt = 0.026  # Tensão térmica

# Definir parâmetros do circuito
Vcc = 5   # Tensão de alimentação
Rc = 470  # Resistência do coletor
Rb = 10   # Resistência da base

# Gerar a variação de tensão base-emissor
Vbe = np.linspace(0, 0.7, 100)

# Calcular a corrente do coletor
Ic = (Vcc - Vbe) / (Rc + (Rb / (1 + (beta * (Vbe / Vt)))))

# Plotar o gráfico I-V
plt.plot(Vbe, Ic, color='red', linestyle='-', linewidth=2)

# Adicionar rótulos de eixo e título
plt.xlabel('Tensão base-emissor (V)')
plt.ylabel('Corrente do coletor (mA)')
plt.title('Gráfico I-V de um transistor BJT')

# Adicionaro grade ao fundo do gráfico
plt.grid(True)

# Exibir o gráfico
plt.show()


```
